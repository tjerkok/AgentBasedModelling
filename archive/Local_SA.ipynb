```python
import matplotlib.pyplot as plt
import numpy as np
from load import local_SA, load_SA
```

```python
data_avg_arrival = load_SA("local_SA_data_avg_arrival.pkl")
data_vision2 = load_SA("local_SA_data_vision2.pkl")
data_vision2_prob = load_SA("local_SA_data_vision2_prob.pkl")
data_familiar2 = load_SA("local_SA_data_familiar2.pkl")
data_familiar2_prob = load_SA("local_SA_data_familiar2_prob.pkl")
data_speed2 = load_SA("local_SA_data_speed2.pkl")
data_speed2_prob = load_SA("local_SA_data_speed2_prob.pkl")
```

```python
print(data_avg_arrival, "\n")
print(data_vision2, "\n")
print(data_vision2_prob, "\n")
print(data_familiar2, "\n")
print(data_familiar2_prob, "\n")
print(data_speed2, "\n")
print(data_speed2_prob, "\n")
```

```python
fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()
fig3, ax3 = plt.subplots()
data = data_avg_arrival["avg_arrival"]
print("avg arrival, interactions, timesteps, distance")
for data_i in data:
    x = data_i[0]
    interactions, time_steps, distance = [data_i[1]["mean_interactions"]], [data_i[1]["mean_time_steps"]], [data_i[1]["mean_distance"]]
    print(x, np.mean(interactions), np.mean(time_steps), np.mean(distance))
    ax1.errorbar(x, np.mean(interactions), yerr=2*np.std(interactions), fmt="k.", ecolor="gray")
    ax2.errorbar(x, np.mean(time_steps), yerr=2*np.std(time_steps), fmt="k.", ecolor="gray")
    ax3.errorbar(x, np.mean(distance), yerr=2*np.std(distance), fmt="k.", ecolor="gray")
    
ax1.set_ylabel("interactions")
ax1.set_xlabel("avg arrival rate")
ax2.set_ylabel("time steps")
ax2.set_xlabel("avg arrival rate")
ax3.set_ylabel("distance")
ax3.set_xlabel("avg arrival rate")
fig1.savefig("figs/avg_arrival_interactions")
fig2.savefig("figs/avg_arrival_timesteps")
fig3.savefig("figs/avg_arrival_distance")
```
